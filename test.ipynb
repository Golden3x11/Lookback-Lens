```python
import pandas as pd
import os

REPO_DIR = '../Research'
DATA_DIR = 'gemma_att'
HALLU_PATH = '/net/pr2/projects/plgrid/plggllmhallu/hallu/'

DS_NAME = 'evaluated_df.parquet'

DATASETS = ['cnndm', 'xsum', 'nq', 'hotpotqa_en']

for data_dir in ['gemma_att', 'llama2_att', 'llama3_att', 'phi_att']:
    ds_name = 'evaluated_df.parquet'

    df = pd.read_parquet(os.path.join(HALLU_PATH, data_dir, DS_NAME)).reset_index(drop=True)

    df = df[df['dataset'].isin(DATASETS)]

    percent_true = (
        df.groupby('dataset')['decision']
        .agg(total_sum='sum', percentage_true=lambda x: round(x.sum() / len(x) * 100, 1))
        .reset_index()
    )

    print(f'# {data_dir.upper()}')
    print(percent_true)
    print()
```
